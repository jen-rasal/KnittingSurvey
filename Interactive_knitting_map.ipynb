```python
df = pd.read_excel (r'C:\Users\jennifer.rasal\OneDrive - Marine Stewardship Council\Documents\Kaggle\knitting\test_results.xlsx')
# create a df with just the country name and lat and long
data = [df["What country do you live in?"], df["Latitude"], df["Longitude"]]
headers = ["Country", "Latitude", "Longitude"]

df3 = pd.concat(data, axis=1, keys=headers)
values = df3['Country'].value_counts()
df_values = df3.drop_duplicates()
```

```python
values = values.to_frame()
values.reset_index(level=0, inplace=True)
values = values.rename(columns={"index": "Country", "Country": "count"})
```

```python
df_values = df_values.sort_values(by=['Country'])
values = values.sort_values(by=['Country'])
df_values = df_values.reset_index()
values = values.reset_index()
d = {'Country': df_values["Country"], 'Latitude': df_values["Latitude"], "Longitude": df_values["Longitude"], "Count": values["count"] }
df = pd.DataFrame(data=d)
df = df.dropna()
```

```python
mymap = folium.Map(
location = [13.133932434766733, 16.103938729508073],
    zoom_start = 2)

for _, Country in df.iterrows():
    folium.Marker(
        location=[Country["Latitude"], Country["Longitude"]],
        popup = Country["Country"] + " " + str(Country["Count"]),
        tooltip = Country["Country"] + " " + str(Country["Count"])
    ).add_to(mymap)
mymap
```

```python
mymap.save("knitting_responses_map.html")
```
